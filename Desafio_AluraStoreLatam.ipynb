### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# suma ventas por tienda
ventas_tienda = tienda['Precio'].sum()
ventas_tienda2 = tienda2['Precio'].sum()
ventas_tienda3 = tienda3['Precio'].sum()
ventas_tienda4 = tienda4['Precio'].sum()

# print de suma por tienda
print(f"Ventas totales - tienda: ${ventas_tienda:,.2f}")
print(f"Ventas totales - tienda2: ${ventas_tienda2:,.2f}")
print(f"Ventas totales - tienda3: ${ventas_tienda3:,.2f}")
print(f"Ventas totales - tienda4: ${ventas_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt

# suma ventas por tienda
facturacion_por_tienda = {
    'tienda': tienda['Precio'].sum(),
    'tienda2': tienda2['Precio'].sum(),
    'tienda3': tienda3['Precio'].sum(),
    'tienda4': tienda4['Precio'].sum()
}


nombres_tiendas = list(facturacion_por_tienda.keys())
montos = list(facturacion_por_tienda.values())

# gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(nombres_tiendas, montos, color='skyblue')

# títulos y etiquetas
plt.title('Facturación Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación (en $)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# valores
for i, v in enumerate(montos):
    plt.text(i, v + max(montos)*0.01, f"${v:,.0f}", ha='center', fontsize=9)


plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Promedio de calificación - tienda: {promedio_tienda:.2f}")
print(f"Promedio de calificación - tienda2: {promedio_tienda2:.2f}")
print(f"Promedio de calificación - tienda3: {promedio_tienda3:.2f}")
print(f"Promedio de calificación - tienda4: {promedio_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos
labels = ['Tienda', 'Tienda2', 'Tienda3', 'Tienda4']
promedios = [promedio_tienda, promedio_tienda2, promedio_tienda3, promedio_tienda4]

# grafico
plt.figure(figsize=(8, 5))
barras = plt.barh(labels, promedios, color='mediumseagreen', alpha=0.8)

# Título y etiquetas
plt.title('Promedio de calificación por tienda')
plt.xlabel('Calificación promedio (1 a 5)')
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.4)

# valores
for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho + 0.05, barra.get_y() + barra.get_height()/2,
             f'{ancho:.2f}', va='center', fontsize=9)

plt.tight_layout()
plt.show()

```

# 4. Productos más y menos vendidos

```python
# Para TIENDA
conteo_tienda = tienda['Producto'].value_counts()
producto_mas_vendido_tienda = conteo_tienda.idxmax()
producto_menos_vendido_tienda = conteo_tienda.idxmin()

print("Tienda")
print(f"Producto más vendido: {producto_mas_vendido_tienda} ({conteo_tienda.max()} ventas)")
print(f"Producto menos vendido: {producto_menos_vendido_tienda} ({conteo_tienda.min()} ventas)\n")

# Para TIENDA2
conteo_tienda2 = tienda2['Producto'].value_counts()
producto_mas_vendido_tienda2 = conteo_tienda2.idxmax()
producto_menos_vendido_tienda2 = conteo_tienda2.idxmin()

print("Tienda2")
print(f"Producto más vendido: {producto_mas_vendido_tienda2} ({conteo_tienda2.max()} ventas)")
print(f"Producto menos vendido: {producto_menos_vendido_tienda2} ({conteo_tienda2.min()} ventas)\n")

# Para TIENDA3
conteo_tienda3 = tienda3['Producto'].value_counts()
producto_mas_vendido_tienda3 = conteo_tienda3.idxmax()
producto_menos_vendido_tienda3 = conteo_tienda3.idxmin()

print("Tienda3")
print(f"Producto más vendido: {producto_mas_vendido_tienda3} ({conteo_tienda3.max()} ventas)")
print(f"Producto menos vendido: {producto_menos_vendido_tienda3} ({conteo_tienda3.min()} ventas)\n")

# Para TIENDA4
conteo_tienda4 = tienda4['Producto'].value_counts()
producto_mas_vendido_tienda4 = conteo_tienda4.idxmax()
producto_menos_vendido_tienda4 = conteo_tienda4.idxmin()

print("Tienda4")
print(f"Producto más vendido: {producto_mas_vendido_tienda4} ({conteo_tienda4.max()} ventas)")
print(f"Producto menos vendido: {producto_menos_vendido_tienda4} ({conteo_tienda4.min()} ventas)")
```

# 5. Envío promedio por tienda

```python

envio_prom_tienda = tienda['Costo de envío'].mean()
envio_prom_tienda2 = tienda2['Costo de envío'].mean()
envio_prom_tienda3 = tienda3['Costo de envío'].mean()
envio_prom_tienda4 = tienda4['Costo de envío'].mean()


print(f'Costo de envío promedio - tienda: ${envio_prom_tienda:,.2f}')
print(f'Costo de envío promedio - tienda2: ${envio_prom_tienda2:,.2f}')
print(f'Costo de envío promedio - tienda3: ${envio_prom_tienda3:,.2f}')
print(f'Costo de envío promedio - tienda4: ${envio_prom_tienda4:,.2f}')
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda', 'Tienda2', 'Tienda3', 'Tienda4']
costos_envio = [envio_prom_tienda, envio_prom_tienda2, envio_prom_tienda3, envio_prom_tienda4]


plt.figure(figsize=(8, 5))
plt.plot(tiendas, costos_envio, marker='o', linestyle='-', color='dodgerblue', linewidth=2)


plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio ($)')
plt.grid(True, linestyle='--', alpha=0.5)


for i, valor in enumerate(costos_envio):
    plt.text(i, valor + 500, f"${valor:,.0f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# 6. Análisis del desempeño geográfico


```python
import matplotlib.pyplot as plt

tabla_tienda_lugar = ventas.groupby(['Tienda', 'Lugar de Compra']).size().reset_index(name='Cantidad de Ventas')
tabla_tienda_lugar = tabla_tienda_lugar.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, False])

tiendas_unicas = tabla_tienda_lugar['Tienda'].unique()

for tienda_nombre in tiendas_unicas:
    datos_tienda = tabla_tienda_lugar[tabla_tienda_lugar['Tienda'] == tienda_nombre]

    plt.figure(figsize=(10, 5))
    barras = plt.bar(datos_tienda['Lugar de Compra'], datos_tienda['Cantidad de Ventas'], color='steelblue')

    plt.title(f'Ventas por ciudad - {tienda_nombre}')
    plt.xlabel('Lugar de Compra')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)

    for barra in barras:
        altura = barra.get_height()
        plt.text(barra.get_x() + barra.get_width() / 2, altura + 3, f'{int(altura)}',
                 ha='center', va='bottom', fontsize=9)

    plt.tight_layout()
    plt.show()
```

# 7. Analisis facturacion - costo de envio

```python
#  suma ventas - costo de envio
tienda['Facturación Neta'] = tienda['Precio'] - tienda['Costo de envío']
tienda2['Facturación Neta'] = tienda2['Precio'] - tienda2['Costo de envío']
tienda3['Facturación Neta'] = tienda3['Precio'] - tienda3['Costo de envío']
tienda4['Facturación Neta'] = tienda4['Precio'] - tienda4['Costo de envío']

#  suma fact neta
neto_tienda = tienda['Facturación Neta'].sum()
neto_tienda2 = tienda2['Facturación Neta'].sum()
neto_tienda3 = tienda3['Facturación Neta'].sum()
neto_tienda4 = tienda4['Facturación Neta'].sum()


print(f"Facturación neta - tienda:  ${neto_tienda:,.2f}")
print(f"Facturación neta - tienda2: ${neto_tienda2:,.2f}")
print(f"Facturación neta - tienda3: ${neto_tienda3:,.2f}")
print(f"Facturación neta - tienda4: ${neto_tienda4:,.2f}")
```
